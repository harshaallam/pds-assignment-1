# Visualizations

### Histogram Plot

```python
plt.figure(figsize=(8,5))
sns.histplot(df['math score'],bins=10,kde=True,color='blue')
plt.title('Distribution of math scores')
plt.xlabel("Math Score")
plt.ylabel("Frequency")
plt.show()
```

**This histogram helps us understand how math scores are distributed among students. The KDE curve shows if the data is normally distributed or skewed**.

### Bar Plot

```python
plt.figure(figsize=(8,5))
df.groupby("gender")[["math score", "reading score", "writing score"]].mean().plot(kind="bar")
plt.title("Average Scores by Gender")
plt.xlabel("Gender")
plt.ylabel("Average Score")
plt.legend(["Math", "Reading", "Writing"])
plt.show()
```

**This bar chart compares average scores between male and female students, helping us see if there are gender-based differences in performance.**

### Box Plot

```python
plt.figure(figsize=(8,5))
sns.boxplot(x="ethnicity", y="writing score", data=df)
plt.title("Writing Scores by Race/Ethnicity")
plt.xlabel("Race/Ethnicity")
plt.ylabel("Writing Score")
plt.show()
```

**This boxplot helps us identify variations in writing scores across different racial/ethnic groups. It also highlights outliers in the data.**

### Scatter Plot

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x="reading score", y="writing score", hue="gender", data=df)
plt.title("Reading Score vs Writing Score")
plt.xlabel("Reading Score")
plt.ylabel("Writing Score")
plt.legend(title="Gender")
plt.show()
```

**The scatter plot helps identify correlations between reading and writing scores. If most points are close to a straight line, it suggests a strong positive correlation.**

### Box Plot

```python
plt.figure(figsize=(8,5))
sns.boxplot(x="test preparation course", y="math score", data=df)
plt.title("Math Scores by Test Preparation Course")
plt.xlabel("Test Preparation Course")
plt.ylabel("Math Score")
plt.show()
```

**This boxplot shows whether students who completed a test preparation course performed better in math compared to those who did not.**
