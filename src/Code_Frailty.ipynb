```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('Assgn1.csv')
df
```

```python
df['Frailty']=df['Frailty'].map({'N':0,'Y':1})
df
```

```python
df.to_csv('cleaned_data.csv')
```

```python
correlation=df.corr()
correlation['Frailty'].sort_values()
```

```python
sns.scatterplot(x=df['Grip Strength'],y=df['Frailty'])
plt.xlabel("Grip Strength (KG)")
plt.ylabel("Frailty (1=yes, 0=No)")
plt.title("Grip Strength vs Frailty")
plt.show()
```

```python
sns.scatterplot(x=df.Weight, y=df['Frailty'])
plt.xlabel("Weight (lbs)")
plt.ylabel("Frailty (1=yes, 0=No)")
plt.title("Weight vs Frailty")
plt.show()
```

```python
sns.scatterplot(x=df.Age, y=df['Frailty'])
plt.xlabel("Age (years)")
plt.ylabel("Frailty (1=yes, 0=No)")
plt.title("Age vs Frailty")
plt.show()
```

```python
sns.scatterplot(x=df.Height, y=df['Frailty'])
plt.xlabel("Height (inches)")
plt.ylabel("Frailty (1=yes, 0=No)")
plt.title("Height vs Frailty")
plt.show()
```
